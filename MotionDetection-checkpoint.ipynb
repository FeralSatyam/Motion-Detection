```python
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
    image.flags.writeable = False                  # Image is no longer writeable
    results = model.process(image)                 # Make prediction
    image.flags.writeable = True                   # Image is now writeable 
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
    return image, results
```

```python
import cv2
import mediapipe as mp
import time
import pygame
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import threading
import random  # Mock random number to simulate calling a number

# Initialize pygame mixer for sound playback
pygame.mixer.init()

# Function to perform MediaPipe detection
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert the image color to RGB
    results = model.process(image)  # Process the image and make detections
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)  # Convert back to BGR for OpenCV
    return image, results

# Function to draw landmarks on the image
def draw_landmarks(image, results):
    mp_drawing = mp.solutions.drawing_utils
    mp_holistic = mp.solutions.holistic
    
    # Draw face landmarks
    if results.face_landmarks:
        mp_drawing.draw_landmarks(
            image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS, 
            mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), 
            mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
        )
    
    # Draw pose landmarks
    if results.pose_landmarks:
        mp_drawing.draw_landmarks(
            image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
            mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
        )
    
    # Draw left hand landmarks
    if results.left_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
            mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
        )
    
    # Draw right hand landmarks
    if results.right_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
            mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
        )

# Start detection in a separate thread to avoid blocking the UI
def start_detection():
    global detection_active
    detection_active = True
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        messagebox.showerror("Error", "Could not open video device.")
        return

    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    out = None
    recording = False
    motion_detected = False
    last_motion_time = time.time()
    sound_played = False

    snooze_start_time = None  # Time when snooze started

    with mp.solutions.holistic.Holistic(min_detection_confidence=0.2, min_tracking_confidence=0.2) as holistic:
        while cap.isOpened() and detection_active:
            ret, frame = cap.read()
            if not ret:
                break

            frame = cv2.flip(frame, 1)
            image, results = mediapipe_detection(frame, holistic)

            motion_detected = results.face_landmarks or results.left_hand_landmarks or results.right_hand_landmarks

            # If motion is detected, play the sound and display the text
            if motion_detected:
                if not sound_played:
                    pygame.mixer.music.load('sound.mp3')
                    pygame.mixer.music.play(loops=-1, start=0.0)  # Loop the sound indefinitely
                    sound_played = True
                    snooze_start_time = time.time()  # Start snooze timer

                if not recording:
                    recording = True
                    timestamp = time.strftime("%Y%m%d-%H%M%S")
                    out = cv2.VideoWriter(f"motion_{timestamp}.avi", cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width, frame_height))
                    last_motion_time = time.time()

                # Display motion detected text
                cv2.putText(image, "Motion Detected!", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 
                            1, (0, 0, 255), 2, cv2.LINE_AA)

            # If no motion detected, stop the sound
            else:
                if sound_played:
                    pygame.mixer.music.stop()  # Stop the sound if no motion
                    sound_played = False

            if recording and out:
                out.write(frame)

            if recording and not motion_detected:
                if time.time() - last_motion_time > 5:
                    recording = False
                    if out:
                        out.release()
                        out = None
                else:
                    last_motion_time = time.time()

            # If snooze has started, check if 10 seconds have passed
            if snooze_start_time and time.time() - snooze_start_time > 10:
                call_number()  # Call the predefined number

            draw_landmarks(image, results)
            cv2image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(cv2image)
            imgtk = ImageTk.PhotoImage(image=img)
            video_label.imgtk = imgtk
            video_label.configure(image=imgtk)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        if out:
            out.release()

        if not detection_active:  # If detection is stopped, stop the sound as well
            pygame.mixer.music.stop()

# Function to start detection when the button is pressed
def on_start_button_click():
    start_button.config(state=tk.DISABLED)  # Disable the start button
    stop_button.config(state=tk.NORMAL)  # Enable the stop button
    snooze_button.config(state=tk.NORMAL)  # Enable the snooze button
    detection_thread = threading.Thread(target=start_detection)
    detection_thread.start()

# Function to stop detection when the button is pressed
def on_stop_button_click():
    global detection_active
    detection_active = False
    stop_button.config(state=tk.DISABLED)  # Disable the stop button
    start_button.config(state=tk.NORMAL)  # Enable the start button
    snooze_button.config(state=tk.DISABLED)  # Disable the snooze button

# Function to snooze the detection temporarily
def on_snooze_button_click():
    global snooze_start_time
    snooze_start_time = time.time()  # Start the snooze timer
    pygame.mixer.music.stop()  # Stop the sound
    messagebox.showinfo("Snooze", "Motion detection is snoozed. The sound is paused.")
    snooze_button.config(state=tk.DISABLED)  # Disable snooze button while snoozed
    stop_snooze_button.config(state=tk.NORMAL)  # Enable the stop snooze button

# Function to stop snoozing and resume detection
def on_stop_snooze_button_click():
    global snooze_start_time
    snooze_start_time = None  # Reset snooze timer
    pygame.mixer.music.play(loops=-1, start=0.0)  # Resume the sound
    snooze_button.config(state=tk.NORMAL)  # Enable snooze button again
    stop_snooze_button.config(state=tk.DISABLED)  # Disable the stop snooze button

# Mock function to simulate calling a predefined number
def call_number():
    phone_number = "+1234567890"  # Example number
    print(f"Calling {phone_number}...")  # Simulating a call

# Set up the main Tkinter window
root = tk.Tk()
root.title("Motion Detection System")

# Create and pack the start button
start_button = tk.Button(root, text="Start Detection", command=on_start_button_click)
start_button.pack(pady=10)

# Create and pack the stop button
stop_button = tk.Button(root, text="Stop Detection", command=on_stop_button_click, state=tk.DISABLED)
stop_button.pack(pady=10)

# Create and pack the snooze button
snooze_button = tk.Button(root, text="Snooze", command=on_snooze_button_click, state=tk.DISABLED)
snooze_button.pack(pady=10)

# Create and pack the stop snooze button
stop_snooze_button = tk.Button(root, text="Stop Snooze", command=on_stop_snooze_button_click, state=tk.DISABLED)
stop_snooze_button.pack(pady=10)

# Create label for video feed display
video_label = tk.Label(root)
video_label.pack()

# Initialize global variables to control detection and snooze
detection_active = False
snooze_start_time = None

# Start the Tkinter main loop
root.mainloop()
```

```python
import cv2
import mediapipe as mp
import time
import pygame
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import threading
import random  # Mock random number to simulate calling a number

# Initialize pygame mixer for sound playback
pygame.mixer.init()

# Function to perform MediaPipe detection
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert the image color to RGB
    results = model.process(image)  # Process the image and make detections
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)  # Convert back to BGR for OpenCV
    return image, results

# Function to draw landmarks on the image
def draw_landmarks(image, results):
    mp_drawing = mp.solutions.drawing_utils
    mp_holistic = mp.solutions.holistic
    
    # Draw face landmarks
    if results.face_landmarks:
        mp_drawing.draw_landmarks(
            image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS, 
            mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), 
            mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
        )
    
    # Draw pose landmarks
    if results.pose_landmarks:
        mp_drawing.draw_landmarks(
            image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
            mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
        )
    
    # Draw left hand landmarks
    if results.left_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
            mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
        )
    
    # Draw right hand landmarks
    if results.right_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
            mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
        )

# Start detection in a separate thread to avoid blocking the UI
def start_detection():
    global detection_active
    detection_active = True
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        messagebox.showerror("Error", "Could not open video device.")
        return

    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    out = None
    recording = False
    motion_detected = False
    last_motion_time = time.time()
    sound_played = False

    snooze_start_time = None  # Time when snooze started
    snooze_mode = False  # New flag to track snooze mode
    continuous_motion_start_time = None  # To track the start of continuous motion
    continuous_motion_detected = False  # Flag to track if continuous motion has been detected for 10 seconds

    with mp.solutions.holistic.Holistic(min_detection_confidence=0.2, min_tracking_confidence=0.2) as holistic:
        while cap.isOpened() and detection_active:
            ret, frame = cap.read()
            if not ret:
                break

            frame = cv2.flip(frame, 1)
            image, results = mediapipe_detection(frame, holistic)

            motion_detected = results.face_landmarks or results.left_hand_landmarks or results.right_hand_landmarks

            # If motion is detected and not in snooze mode, play the sound
            if motion_detected and not snooze_mode:
                if not sound_played:  # Only play sound if not already playing
                    pygame.mixer.music.load('sound.mp3')
                    pygame.mixer.music.play(loops=-1, start=0.0)  # Loop the sound indefinitely
                    sound_played = True

                if not recording:
                    recording = True
                    timestamp = time.strftime("%Y%m%d-%H%M%S")
                    out = cv2.VideoWriter(f"motion_{timestamp}.avi", cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width, frame_height))
                    last_motion_time = time.time()

                # Check if motion has been continuous for 10 seconds
                if continuous_motion_start_time is None:
                    continuous_motion_start_time = time.time()  # Start the 10-second timer
                elif time.time() - continuous_motion_start_time > 10:
                    continuous_motion_detected = True

                # Display motion detected text
                cv2.putText(image, "Motion Detected!", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 
                            1, (0, 0, 255), 2, cv2.LINE_AA)

            # If no motion detected and the sound is playing, stop the sound
            elif not motion_detected and sound_played:
                pygame.mixer.music.stop()  # Stop the sound if no motion
                sound_played = False

                # Reset continuous motion tracking when motion stops
                continuous_motion_start_time = None
                continuous_motion_detected = False

            # If continuous motion has been detected for 10 seconds and snooze mode is active, display the call message
            if continuous_motion_detected and snooze_mode:
                cv2.putText(image, "Calling...", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 
                            1, (0, 255, 0), 2, cv2.LINE_AA)

            if recording and out:
                out.write(frame)

            if recording and not motion_detected:
                if time.time() - last_motion_time > 5:
                    recording = False
                    if out:
                        out.release()
                        out = None
                else:
                    last_motion_time = time.time()

            # If snooze has started, check if 10 seconds have passed
            if snooze_start_time and time.time() - snooze_start_time > 10:
                call_number()  # Call the predefined number

            draw_landmarks(image, results)
            cv2image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(cv2image)
            imgtk = ImageTk.PhotoImage(image=img)
            video_label.imgtk = imgtk
            video_label.configure(image=imgtk)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        if out:
            out.release()

        if not detection_active:  # If detection is stopped, stop the sound as well
            pygame.mixer.music.stop()

# Function to start detection when the button is pressed
def on_start_button_click():
    start_button.config(state=tk.DISABLED)  # Disable the start button
    stop_button.config(state=tk.NORMAL)  # Enable the stop button
    snooze_button.config(state=tk.NORMAL)  # Enable the snooze button
    detection_thread = threading.Thread(target=start_detection)
    detection_thread.start()

# Function to stop detection when the button is pressed
def on_stop_button_click():
    global detection_active
    detection_active = False
    stop_button.config(state=tk.DISABLED)  # Disable the stop button
    start_button.config(state=tk.NORMAL)  # Enable the start button
    snooze_button.config(state=tk.DISABLED)  # Disable the snooze button

# Function to snooze the detection temporarily
def on_snooze_button_click():
    global snooze_start_time, snooze_mode
    snooze_start_time = time.time()  # Start the snooze timer
    snooze_mode = True  # Enable snooze mode
    pygame.mixer.music.stop()  # Stop the sound
    messagebox.showinfo("Snooze", "Motion detection is snoozed. The sound is paused.")
    snooze_button.config(state=tk.DISABLED)  # Disable snooze button while snoozed
    stop_snooze_button.config(state=tk.NORMAL)  # Enable the stop snooze button

# Function to stop snoozing and resume detection
def on_stop_snooze_button_click():
    global snooze_start_time, snooze_mode
    snooze_start_time = None  # Reset snooze timer
    snooze_mode = False  # Disable snooze mode
    snooze_button.config(state=tk.NORMAL)  # Enable snooze button again
    stop_snooze_button.config(state=tk.DISABLED)  # Disable the stop snooze button

# Mock function to simulate calling a predefined number
def call_number():
    phone_number = "+1234567890"  # Example number
    print(f"Calling {phone_number}...")  # Simulating a call

# Set up the main Tkinter window
root = tk.Tk()
root.title("Motion Detection System")

# Create and pack the start button
start_button = tk.Button(root, text="Start Detection", command=on_start_button_click)
start_button.pack(pady=10)

# Create and pack the stop button
stop_button = tk.Button(root, text="Stop Detection", command=on_stop_button_click, state=tk.DISABLED)
stop_button.pack(pady=10)

# Create and pack the snooze button
snooze_button = tk.Button(root, text="Snooze", command=on_snooze_button_click, state=tk.DISABLED)
snooze_button.pack(pady=10)

# Create and pack the stop snooze button
stop_snooze_button = tk.Button(root, text="Stop Snooze", command=on_stop_snooze_button_click, state=tk.DISABLED)
stop_snooze_button.pack(pady=10)

# Create label for video feed display
video_label = tk.Label(root)
video_label.pack()

# Initialize global variables to control detection and snooze
detection_active = False
snooze_start_time = None
snooze_mode = False  # Flag to track snooze mode

# Start the Tkinter main loop
root.mainloop()
```
